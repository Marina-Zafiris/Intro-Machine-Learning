# K- Means Clustering

K-Means clustering algorithm is an unsupervised learning that has an learning iterative process in which the dataset are grouped into k numbers of predefined non-overlapping clusters. The sum of the squared distance between the clusters centroid and the data point is at a minimum, at this position the centroid of the cluster is the arithmetic mean of the data points that are in the clusters.

1. Randomly initialize the cluster centers of each cluster
2. For each data point, compute the euclidian distance from all the centroids
3. Adjust the centroid of each cluster by taking the average of all the data 

```python
# Necessary Packages
import numpy as np
import pandas as pd
import random as rd
from math import sqrt
import matplotlib.pyplot as plt
from copy import deepcopy
```

We will be assigning each data point to a centroid based off of the minimun distance it has from the colection of possible centroid.

$\underset{c_i \in C}{\arg\min} \; dist(c_i,x)^2$

Through each each iteration we go through in N, the centroids are recomputed. 

```python
# Import the data set
dataset=pd.read_csv('Mall_Customers.csv')
dataset.describe()

# Reducing it to 2 column for later visualization purposes
X = dataset.iloc[:, [3, 4]].values # Notice that we are not selecting a Y or predefined output, unsupervised

# X is an mxn matrix
m=X.shape[0] #number of training examples
n=X.shape[1] #number of features

Num=200

K = 5 # Number of clusters we want

# Initialize an empty centroids array
Centroids=np.array([]).reshape(n,0) 

# Selecting random centroids
for i in range(K):
    rand=rd.randint(0,m-1) # Chooses random index number for length of points
    Centroids=np.c_[Centroids,X[rand]] # There will be K stacked points along the columns 

Old_Centroids = deepcopy(Centroids) # Storing the centroid so we can plot the change in the centroids

for i in range(Num): # Looping through our 200 iterations
     
    Distance=np.array([]).reshape(m,0) #Initializing empty array for distance for each point
    
    for k in range(K): #looping through our number of desired clusters
        temp_dist=np.sum((X-Centroids[:,k])**2,axis=1) 
        Distance=np.c_[Distance,temp_dist] # Stacks them along the column
    C_Num=np.argmin(Distance,axis=1)+1 # +1 because python starts at 0, cluster number
    
    #Storing the cluster point pairs in a dictionary, where cluster # is the key and values are the x points
    #We are creating a Y output because we dont know exactly how they should be classified
    Y={}
    for k in range(K):
        Y[k+1]=np.array([]).reshape(2,0)
    for i in range(m):
        Y[C_Num[i]]=np.c_[Y[C_Num[i]],X[i]]
     
    for k in range(K):
        Y[k+1]=Y[k+1].T
    
    for k in range(K):
        Centroids[:,k]=np.mean(Y[k+1],axis=0)
```

### Plotting

```python
color=['tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan']
labels=['C 1','C 2','C 3','C 4','C 5']

plt.figure(figsize =(14, 12))

for k in range(K):
    plt.scatter(Y[k+1][:,0],Y[k+1][:,1],c=color[k],label=labels[k])

plt.scatter(Old_Centroids[0,:],Old_Centroids[1,:],s=200,c='red',label='Old Centroids')
plt.scatter(Centroids[0,:],Centroids[1,:],s=200,c='lime',label='New Centroids')

plt.xlabel('Income') # Info from the original data set
plt.ylabel('Number of transactions')
plt.legend()
plt.show()
```
